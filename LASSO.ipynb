```python
y = train['income_morethan_50K']
```

```python
X_numerical = train.drop(['income_morethan_50K','ACLSWKR', 'AHGA','AHSCOL','AMARITL','AMJIND','AMJOCC', 'ARACE', 'AREORGN',
                               'ASEX','AUNMEM', 'AUNTYPE','AWKSTAT','FEDTAX','FILESTAT','GRINREG','GRINST',
                               'HHDFMX','MIGMTR1','MIGMTR3','MIGMTR4','MIGSAME','PARENT','PEFNTVTY','PEMNTVTY',
                               'PENATVTY','PRCITSHP','VETQVA'], axis=1).astype('float64')
```

```python
list_numerical = X_numerical.columns
list_numerical
```

```python
print(dummies.columns)
```

```python
X = pd.concat([X_numerical, dummies[['ACLSWKR_Federal government', 'ACLSWKR_Local government',
       'ACLSWKR_Never worked', 'ACLSWKR_Not in universe', 'ACLSWKR_Private',
       'ACLSWKR_Self-employed-incorporated',
       'ACLSWKR_Self-employed-not incorporated', 'ACLSWKR_State government',
       'ACLSWKR_Without pay', 'AHGA_10th grade']]], axis=1)
X.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```
